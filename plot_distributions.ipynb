# Plots of Demo Distributions

This notebook generates histograms of each distribution in our library for visual inspection.

```python
import random_distributions as rds
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Import the other script where we define all the functions
```

## Test LCG - should make Uniform(0,1)


```python
# Instantiate lcg class to make the object
lcg = rds.lcg_class()

unifs = [lcg.rand() for _ in range(100_000)]

print(f"Min Random is {min(unifs)}")
print(f"Max Random is {max(unifs)}")

plt.hist(unifs, bins=20)
plt.title(f"Sample random uniform between 0 and 1 from LCG")
plt.ylabel("Number of Observations")
plt.show()
```

## Uniform(a,b) Distribution

```python
a = 3
b = 6
unifs = [rds.unif(a, b) for _ in range(100_000)]
# print(unifs)

plt.hist(unifs, bins=20)
plt.title(f"Sample uniform distributions between {a} and {b}")
plt.ylabel("Number of Observations")
plt.show()
```

## Bernoulli(p) Distribution

```python
p = 0.2
berns = [rds.bern(p) for i in range(1_000)]
plt.hist(berns, bins=2)
plt.title(f"Sample Bernoulli distribution with probability {p}")
plt.ylabel("Number of Observations")
plt.show()
```

## Binomial(n, p) Distribution

```python
n = 10
p = 0.3

binoms = [rds.binom(n, p) for _ in range(10_000)]
print(f"Max Binomial is {max(binoms)}")
print(f"Min Binomial is {min(binoms)}")
plt.hist(binoms, bins=max(binoms) - min(binoms)+1)
plt.title(f"Sample Binomial distribution with {n} trials with probability {p}")
plt.ylabel("Number of Observations")
plt.show()
```

```python
n = 15
p = 0.1

binoms = [rds.binom(n, p) for _ in range(10_000)]
print(f"Max Binomial is {max(binoms)}")
print(f"Min Binomial is {min(binoms)}")
plt.hist(binoms, bins=max(binoms) - min(binoms)+1)
plt.title(f"Sample Binomial distribution with {n} trials with probability {p}")
plt.ylabel("Number of Observations")
plt.show()
```

This next distribution should be approximately normal as the binomial distribution approaches normal for large n... please be patient as this one can take a while to make.

```python
n = 1_000
p = 0.45

binoms = [rds.binom(n, p) for _ in range(100_000)]
print(f"Max Binomial is {max(binoms)}")
print(f"Min Binomial is {min(binoms)}")
plt.hist(binoms, bins=max(binoms) - min(binoms)+1)
plt.title(f"Sample Binomial distribution with {n} trials with probability {p}")
plt.ylabel("Number of Observations")
plt.show()
```

## Normal(mu, sigma) Distribution

```python
mu = 5
sig = 3

norms = [rds.norm(mu, sig) for _ in range(100_000)]
plt.hist(norms, bins=100)
plt.title(f"Sample Normal distribution with mean {mu} and standard deviation {sig}")
plt.ylabel("Number of Observations")
plt.show()
```

```python
mu = 40
sig = 80

norms = [rds.norm(mu, sig) for _ in range(100_000)]
plt.hist(norms, bins=100)
plt.title(f"Sample Normal distribution with mean {mu} and standard deviation {sig}")
plt.ylabel("Number of Observations")
plt.show()
```

## Triangular distribution from a to c


```python
a = 3
c = 7

triags = [rds.tria(a, c) for _ in range(100_000)]

print(f"Minimum value in distribution is {min(triags)}")
print(f"Maximum value in distribution is {max(triags)}")


plt.hist(triags, bins=100)
plt.title(f"Sample Triangular Distribution from {a} to {c}")
plt.ylabel("Number of Observations")
plt.show()
```

## Exponential(lambda) distribution


```python
lamb = 3

expos = [rds.expo(lamb) for _ in range(100_000)]
plt.hist(expos, bins=100)
plt.title(f"Sample Exponential({lamb}) Distribution")
plt.ylabel("Number of Observations")
plt.show()
```

## Poisson(lambda) Distribution

```python
lamb = 5

fishies = [rds.poisson(lamb) for _ in range(100_000)]
plt.hist(fishies, bins=50)
plt.title(f"Sample Poisson({lamb}) Distribution")
plt.ylabel("Number of Observations")
plt.show()
```
